```python
questions = ["1.How many entries are there in the release_year column?",
             "2. What is the average release year of the titles?",
             "3. What is the earliest and latest release year recorded in the dataset",
             "4. What does the 25th percentile (1998) indicate about the dataset?",
             "5. What is the significance of the median (50th percentile) value in this dataset?",
             "6. What does the standard deviation of 22.13 tell us about the release years?",
             "7. How much variation exists between the oldest and the most recent release year in the dataset?"]

questions
```

```python
#1. How many entries are there in the release_year column?

#getting the values

df['release_year'].value_counts()
```

```python
# Plotting the frequency of each release year
df['release_year'].value_counts().sort_index().plot(kind='bar', color="skyblue", figsize=(12, 6))
plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Count of Titles")
plt.show()
```

```python
# 2. What is the average release year of the titles?

# Calculate the average release year
average_release_year = df['release_year'].mean()
print(f"The average release year of the titles is {average_release_year:.2f}")
```

```python
# Plot a histogram of the release years
plt.figure(figsize=(8, 6))
plt.hist(df['release_year'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
```

```python
# In the above chart x axis represent time periods in which titles were released and y axis represent frequency of titles.
#Most titles were released after the year 2000, with a significant spike in the 2010s and 2020s. 
#This could suggest a growing trend in the creation or inclusion of modern content in the dataset.
```

```python
# Question 3 
# 3. What is the earliest and latest release year recorded in the dataset

earliest_year = df['release_year'].min()
latest_year = df['release_year'].max()

print(f"The earliest release year is: {earliest_year}")
print(f"The latest release year is: {latest_year}")
```

```python
# Plotting the data
years = [earliest_year, latest_year]
counts = [df[df['release_year'] == earliest_year].shape[0], 
          df[df['release_year'] == latest_year].shape[0]]  # Counts of entries for these years

plt.bar(years, counts, color=['blue', 'green'])
plt.title("Earliest and Latest Release Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.xticks(years)  # Ensures only the specific years are shown on the x-axis
plt.show()
```

```python
# Question 4
#4 What does the 25th percentile (1998) indicate about the dataset?

# Finding the 25th percentile
percentile_25 = df['release_year'].quantile(0.25)
print(f"The 25th percentile of the release_year column is: {percentile_25}")
```

```python
# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.hist(df['release_year'], bins=20, color='skyblue', edgecolor='black')
plt.axvline(percentile_25, color='red', linestyle='dashed', linewidth=2, label=f"25th Percentile (1998)")

# Adding labels and title
plt.title("Distribution of Release Years with 25th Percentile Highlighted")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# This code will generate a histogram of release_year with a vertical dashed line highlighting the 25th percentile value (1998).
# It indicates that 25% of the titles in the dataset were released in 1998 or earlier, reflecting older content in the datase
```

```python
# Question 5 

# What is the significance of the median (50th percentile) value in this dataset?

# Finding the median (50th percentile)
median_value = df['release_year'].median()  # or use quantile(0.5)
print(f"The median (50th percentile) of the release_year column is: {median_value}")
```

```python
# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.hist(df['release_year'], bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(median_value, color='blue', linestyle='dashed', linewidth=2, label=f"Median (50th Percentile): {median_value}")

# Adding labels, title, and legend
plt.title("Distribution of Release Years with Median Highlighted")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# The median indicates that 50% of the titles in the dataset were released in 2011 or earlier, while the remaining 50% were released after 2011
```

```python
# question 6 
#What does the standard deviation of 22.13 tell us about the release years? 

# Calculating the standard deviation of the release_year column
std_dev = df['release_year'].std()
print(f"The standard deviation of the release_year column is: {std_dev}")
```

```python
#A standard deviation of 22.13 means that the release years tend to vary from the average release year (2002.35) by approximately 22 years.
```

```python
# question 7 

# How much variation exists between the oldest and the most recent release year in the dataset?

# Finding the oldest and most recent release years
min_year = df['release_year'].min()
max_year = df['release_year'].max()

# Calculating the range (variation)
year_range = max_year - min_year

print(f"The oldest release year is: {min_year}")
print(f"The most recent release year is: {max_year}")
print(f"The variation between the oldest and most recent release year is: {year_range} years")
```
